CREATING A GEODATAFRAME FROM A CSV FILE

```python
import pandas as pd
csv = pd.read_csv(r'D:\Geospatial Data\all_country_coordinates.csv')
```

```python
csv.head()
```

```python
df = gpd.GeoDataFrame('')
```

```python
gs = GeoSeries([csv.latitude[0],csv.longitude[0]])
```

```python
csv.latitude,csv.longitude
```

```python
csv.latitude[0]
```

```python
coordinates = []
for i in range (0, len(csv['latitude'])):
    a = Point(csv.longitude[i],csv.latitude[i])
    coordinates.append(a)
```

```python
coordinates
```

```python
Points = pd.DataFrame(coordinates,columns = ['Points'])
```

```python
Points.set_geometry(['Points'])
```

```python
Points
```

```python
gdf = gpd.GeoDataFrame(csv, geometry = Points)
```

```python
csv
```

```python
gdf = gpd.GeoDataFrame(csv)
```

```python
gdf['Coordinates'] = Points
```

```python
gdf.head()
```

```python
gdf.set_geometry('Coordinates')
```

```python
gdf.geometry.name
```

```python
gdf.plot(gdf.Coordinates)
```
