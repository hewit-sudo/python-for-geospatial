```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.arange(1,20,0.2)
y = np.sin(x/2)
z = np.cos(x/2)
k = np.cos((x**2)/2)
```

Explicitly creating the figure and axes

```python
fig, ax = plt.subplots()
ax.plot(x,y, label = 'Linear')
```

```python
plt.plot(y)
plt.plot(z)
plt.plot(k)
```

```python
plt.subplot()
plt.plot(y)
plt.subplot(1,3,2)
plt.plot(z)
plt.subplot(1,3,3)
plt.plot(k)
```
